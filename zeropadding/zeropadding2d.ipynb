```python
import tensorflow as tf
```

```python
import numpy as np 
import os
import glob
import cv2
import matplotlib.pyplot as plt
```

```python
fp = '/home/wonchul/mnt/HDD/datasets/rps/train/paper'
```

```python
img_files = glob.glob(os.path.join(fp, "*.png"))
```

```python
img = cv2.imread(img_files[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(type(img), img.shape)
```

```python
plt.imshow(img)
```

```python
top_pad = 10
bottom_pad = 10
left_pad = 10
right_pad = 10
```

```python
inputs = tf.keras.layers.Input(shape=img.shape)
x = inputs
x = tf.keras.layers.ZeroPadding2D(padding=((top_pad, bottom_pad), \
                                           (left_pad, right_pad)))(x)
outputs = x
```

```python
model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
out = model(img[np.newaxis, ...])
```

```python
plt.imshow(out[0].numpy().astype(np.uint8))
```
